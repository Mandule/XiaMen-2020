```python
import os
import pandas as pd
import numpy as np
```

# 1. aum

```python
aum_fils = os.listdir('aum')
aum = []
for f in aum_fils:
    if 'aum' not in f:
        continue
    print(f)
    mon = int((f.split('.')[0]).split('_')[-1].replace('m', ''))
    tmp = pd.read_csv('aum/'+f)
    if mon>=7:
        tmp['year'] = 2019
        tmp['month'] = mon
    else:
        tmp['year'] = 2020
        tmp['month'] = mon
    aum.append(tmp)
aum = pd.concat(aum, axis=0, ignore_index=True)
```

```python
aum.head()
```

```python
aum.to_pickle('aum_month.pkl')
```

# 4. event

```python
event_fils = os.listdir('bigevent')
event = []
for f in event_fils:
    if 'event' not in f:
        continue
    print(f)
    season = int((f.split('.')[0]).split('_')[-1].replace('Q', ''))
    tmp = pd.read_csv('bigevent/'+f)
    if season >=3:
        tmp['year'] = 2019
        tmp['season'] = season
    else:
        tmp['year'] = 2020
        tmp['season'] = season
    event.append(tmp)
event = pd.concat(event, axis=0, ignore_index=True)
```

```python
event.head()
```

```python
for f in ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10', 'E11', 'E12', 'E13', 'E14', 'E16', 'E18']:
    event[f] = pd.to_datetime(event[f])
```

```python
event.to_pickle('event_season.pkl')
```
